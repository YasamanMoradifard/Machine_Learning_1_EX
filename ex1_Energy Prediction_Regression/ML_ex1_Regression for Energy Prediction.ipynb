# 4- Splitting of the data set 

## 4.1 Seperation features and target variable

```python
# We separate the features (axis, feed, distance) and store it in 'X_multi' then we store our target varible (energy) in 'Y_target' to train our model.
X_multi = data.drop('Energy_Requirement', 1)

# energy requirement is our target variable
Y_target = data.Energy_Requirement
```

## 4.2 Splitting the data into training and test sets

```python
# The given dataset is divided into training and test datasets.
import sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X_multi, Y_target, random_state=42)
```

```python
# Checking the shapes of the datasets so that we dont wrongly fit the data
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

><font color='black'>**Question:** How is the default setting of the `train_test_split()` method regarding the distribution of the training and test dataset? Use the documentation for this method! And what are advantages/disadvantages of different distributions (big training dataset, small test dataset or same size of the data sets?</font>
<br>

>**Your Answer:** <span style="color:green"> The `train_test_split()` allowed inputs are lists, numpy arrays, scipy-sparse matrices or pandas dataframes. The following arguments could be given:
- test_size: as default it's None, means it is 0.25 percent of the whole data set.
- train_size: as default it's None, the value is automatically set to the complement of the test size. 
- random_state: as default it's None, use the global random state instance from numpy.random. Calling the function multiple times will reuse the same instance, and will produce different results.
- Shuffle: as default it's True, means shuffle data set before spllitting.
- stratify: as default it's None, means each set contains approximately the same percentage of samples of each target class as the complete set.

>In case of splitting data in train and test set with the portion of 50/50 (same size of the data sets):
- disadvantages: It may cause underfitting. underfitting happens whenever the model couldn't figure out the relation between feature and output correctly and we may see low performance on both test and train sets, Specially in small data sets.</span>
- advantages: In large data sets, we can have faster training procedure.

>splitting data in train and test set with the portion of 20/80:
- disadvantages: it may cause overfitting, specially in large data sets. We can find it regarding the good performance on training data set, but a poor result on the test set.
- advantages: Avoid underfitting. We can make sure that the amount of data for learning statistical realtion between data points is enough.

# 5- Linear regression

## 5.1- Import the library

```python
from sklearn.linear_model import LinearRegression
```

## 5.2- Training of the model

```python
lreg = LinearRegression()
lreg.fit(X_train, Y_train)
```

## 5.3- Prediction

><font color='black'>**Question:** Discuss the used error metrics and the results!
<br></font>

>**Your Answer:** <span style="color:green">To evaluate Regression models' performance, we use Mean Absolute Error (MAE) and Mean Square Error (MSE): 
- **MAE**: is a measure of errors between paired observations expressing the same phenomenon.
- **MSE**: measures the average of the squares of the errorsthat is, the average squared difference between the estimated values and the actual value.

>So, we can say MAE is the exact difference between the grandtruth and the predicted value of a data point, but MSE is the average of square difference between the real and the predicted output.

>MSE and MAE of our regression model indicates that the error is low in both training and test.Additionally, we don't see huge difference the train and test errors. As a result, our model doesn't overfit or underfit.
</span>

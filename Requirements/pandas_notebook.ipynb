## 3)DataFrame Methods

When you upload a dataset for the first time, there are couple of methods that can help you to understand your data better.
For example:
* head()
* tail()
* describe()

```python
df.head()
#head() method shows you the first 5 line of the dataset
```

```python
#by default, head() method shows you first 5 lines. However, you can change it as well. E.g.
df.head(15)
```

```python
df.tail() #tail() is very similar to head(). But it shows you the last 5 elements
```

```python
#by the same token, you can give tail method a parameter as well
df.tail(10)
```

Another useful method for a dataframe, is describe(). Describe() shows you brief statisctical information about the dataset. It only considers the columns with numerical values.

```python
df.describe() 
```

```python
df.index #shows you the indexing information about the dataset
```

```python
 df.columns # this one shows you the columns of the dataset
```

```python
#you can select individual columns
df["PassengerId"]
```

```python
#or you can select multiple columns
df[["PassengerId", "Fare"]]
```

You can check how many of a specific value exists in a column. For example:

```python
df["Sex"].value_counts() #this shows us how many members of genders exist in the dataset
```

```python
df.sort_values(by='Age') #you can also sort the dataframe based on columns
```
